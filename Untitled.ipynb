```python
import pandas as pd
```

```python
from facebook_scraper import get_posts

url = []
text = []
date = []
comments = []
likes = []
user = []

interests = ['סנטר','צפון ישן','הצפון הישן','שלושה חדרים','3 חדרים','שני שותפים','2 שותפים','בוגרשוב','כיכר','לב העיר','רבין', 'פרישמן','גורדון']
ignore = ['מכירה','למכירה']

groupNames = ['287564448778602','telavivrentals','ApartmentsTelAviv','101875683484689','telavivroommates','35819517694'
              ,'458499457501175','dirotTA','2092819334342645','2065942027026008','emptyLiveableTLVapartments','1196843027043598'
             ,'733810383372996','676960712363338','485208978351445','453182374868551','429827780505313','TelAvivApartments']

for i in groupNames:
    for post in get_posts(group=i, pages=6 ,options={"posts_per_page": 50, "allow_extra_requests": False}):
        if any(words in post['post_text'] for words in interests):
            if any(words in post['post_text'] for words in ignore):
                continue
            text.append(post['post_text'][:200])
            url.append(post['post_url'])
            date.append(post['time'])
            comments.append(post['comments'])
            likes.append(post['likes'])
            user.append(post['username'])


data = {
    'date': date,
    'link': url,
    'info': text,
    'username': user,
    '#comments': comments,
    '#likes': likes
}
df = pd.DataFrame.from_dict(data, orient='index')
df = df.transpose()
df
```

```python
from datetime import datetime, timedelta
today = pd.to_datetime(datetime.now())
threeDaysAgo =  pd.to_datetime(datetime.today() - timedelta(days=3))
df = df.loc[(df['date']>threeDaysAgo) & (df['date']<datetime.now())]
df = df.sort_values('date',ascending=False)
df.dropna
df.drop_duplicates(subset = 'info')
df
```

```python
df.to_csv('./results.csv',index=False, encoding = 'utf-8-sig')
```
